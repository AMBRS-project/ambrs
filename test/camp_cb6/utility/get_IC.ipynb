```python
soa = 0.
soag = 0.
soa_defn = pd.read_csv('/home/dquevedo/AMBRS/jupyter_notebooks/soa_defn.csv')
soa_species = list(soa_defn.loc[soa_defn['POA'].str.strip()=='F','PM Name'].str.strip(" '"))
soag_species = soa_defn.loc[soa_defn['POA'].str.strip()=='F','Vapor Name'].str.strip(" '")
soag_species.loc[soa_defn['Rxn Cntr Name'].str.strip(" '").str.len() > 0] = soa_defn.loc[soa_defn['Rxn Cntr Name'].str.strip(" '").str.len() > 0,'Rxn Cntr Name'].str.strip(" '")
soag_species = list(soag_species)

str_dat = ''''''
TSTEP = 12
for i in range(df_gc.size):
    label = df_gc.iloc[i,0]
    if label not in ds.attrs['VAR-LIST']:
        temp_label = soa_defn[soa_defn['Rxn Cntr Name'].str.contains(label)]
        if temp_label.index.size < 1:
            ic = 0.
            str_dat += \
f'''
{label:>16}, {ic:.4e}
'''
            continue
        labels = temp_label['Vapor Name']
    else:
        labels = [label]
    ic = sum([float(ds.isel(COL=70, ROW=64, LAY=0, TSTEP=TSTEP)[lab.strip(" '")]) for lab in labels if lab.strip(" '") in ds.attrs['VAR-LIST']])
    if label in soag_species:
        soag += ic
        continue
    str_dat += \
f'''
{label:>16}, {ic:.4e}
'''

for j in range(df_ae.size):
    label = df_ae.iloc[j,0]
    labels = [lab for lab in ds.attrs['VAR-LIST'].split() if lab[:-1]==label or lab[:-3]==label or lab[:-4]==label or lab==label]
    ic = 1.e-9*sum([float(ds.isel(COL=70, ROW=64, LAY=0, TSTEP=TSTEP)[lab]) for lab in labels])
    if label in soa_species:
        soa += ic
        continue
    str_dat += \
f'''
{label:>16}, {ic:.4e}
'''

for k in range(df_nr.size):
    label = df_nr.iloc[k,0]
    ic = float(ds.isel(COL=70, ROW=64, LAY=0, TSTEP=TSTEP)[label])
    str_dat += \
f'''
{label:>16}, {ic:.4e}
'''

for ii in range(df_constants.size):
    label = df_constants.iloc[ii,0]
    if label == 'M':
        ic =  1.0e+6
    elif label == 'H2':
        ic = 0.56
    elif label == 'N2':
        ic =  0.7808e+6
    elif label == 'O2':
        ic = 0.2095e+6
    elif label == 'CH4':
        ic = 1.85
    else:
        ic = 0.
    str_dat += \
f'''
{label:>16}, {ic:.4e}
'''

soa_mode_merged = pd.Series(ds.attrs['VAR-LIST'].split(), dtype='string')
soa = 1.e-9*sum([ds.isel(COL=70, ROW=64, LAY=0, TSTEP=TSTEP)[list(soa_mode_merged.loc[soa_mode_merged.str.contains('AMT1')])].to_pandas().sum() for lab in soa_species if lab in ds.attrs['VAR-LIST']])
soag = sum([float(ds.isel(COL=70, ROW=64, LAY=0, TSTEP=TSTEP)[lab]) for lab in soag_species if lab in ds.attrs['VAR-LIST'] and len(lab) > 0])

for jj in range(df_other.size):
    label = df_other.iloc[jj,0]
    if label == 'H2O':
        a = 1. - 373.15 / 293.
        a = (((-0.1299 * a - 0.6445) * a - 1.976) * a + 13.3185) * a
        water_vp = 101325.0 * np.exp(a)
        ic = 0.9 * water_vp * 1.0e6 \
                / 101325.0
    elif label == 'SOAG':
        ic = soag
    elif label == 'SOA':
        ic = soa
    else:
        ic = 0.
    str_dat += \
f'''
{label:>16}, {ic:.4e}
'''

print(str_dat)
```

```python
with open('/home/dquevedo/AMBRS/ambrs_mam4_cb6r5_ae7_aq/tests/ic_sulfate_condensation.dat','w') as f:
    f.write(str_dat)
```
